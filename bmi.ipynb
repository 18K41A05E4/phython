```
w=float(input("Enter your weight in pounds:"))
h=float(input("Enter your height in inches:"))
w=w*(0.45359237)
h=h*(0.0254)
bmi=(w/(h*h))
print("Body Mass Index:",bmi)
if bmi<18.5:
  print("Under weight")
elif bmi>=18.5 and bmi<25.0:
  print("Normal")
elif bmi>=25.0 and bmi<30.0:
  print("Over weight")
else:
  print("Obese")
```
