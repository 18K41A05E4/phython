```
year = int(input("Enter year: ") )
res=(year % 12)
print(res)
if res==0: print("monkey")
elif res==1: print("rooster")
elif res==2: print("dog")
elif res==3: print("pig")
elif res==4: print("rat")
elif res==5: print("ox")
elif res==6: print("tiger")
elif res==7: print("rabbit")
elif res==8: print("dragon")
elif res==9: print("snake")
elif res==10: print("horse")
elif res==11: print("sheep")
```
