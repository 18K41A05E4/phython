```
m=str(input("enter a month:"))
if m.lower()=='december' or m.lower()=='january' or m.lower()=='february':
  print("Winter")
elif m.lower()=='march' or m.lower()=='april' or m.lower()=='may':
  print("Spring")
elif m.lower()=='june' or m.lower()=='july' or m.lower()=='august':
  print("Summer")
else:
  print("Autumn")
```
