```
dis= int(input("Distance to be travel: ") )
wt= int(input("Weight of the goods: ") )
if dis>=500:
  if wt>=100:
    print("Amount to be charged:",(dis*5))
  elif wt>=10 and wt<100:
    print("Amount to be charged:",(dis*6))
  elif wt<10:
    print("Amount to be charged:",(dis*7))
else:
  if wt>=100:
    print("Amount to be charged:",(dis*8))
  elif wt<100:
    print("Amount to be charged:",(dis*5))
```
