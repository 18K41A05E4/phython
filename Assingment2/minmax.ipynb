```
"""9. Sort 3 integers
Given three integers (given through user input), sort the numbers using 
|min| and |max| functions."""

a=int(input("Enter a number:"))
b=int(input("Enter b number:"))
c=int(input("Enter c number:"))
x=min(a,b,c)
y=max(a,b,c)
z=(a+b+c)-x-y
print("Sorted order is:",x,z,y)
```
